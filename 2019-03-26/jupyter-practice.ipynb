## Plot

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

x=np.linspace(-3,3,50) # -3 至 3, 平均取 50 點
y1=3*x+3
y2=x**2
plt.plot(x,y1,'--r',label='y1') # ;收尾可隱藏output line記憶體位置
# plt.figure() #新圖新畫布
plt.plot(x,y2,'oc',label='y2')
x0=1
y0=6 # =3*x0+3
plt.scatter(x0,y0,s=50,color='b')
plt.legend(loc='best')
```

## 補充

```python
n=100
np.random.seed(10)
x=np.random.normal(0,1,n)
y=np.random.random(n)*5
plt.figure(figsize=(12,8)) #畫布大小
T=np.random.randint(1,5,n) #製造群組
area=(30*np.random.rand(n,1))**2 #資料大小
plt.grid()
# plt.scatter(x,y)
plt.scatter(x,y,s=area,c=T,cmap='Set1',alpha=.5) #c為群體,s為大小,cmap色系,alpha透明度
plt.colorbar()
plt.savefig(r'C:\Users\user\Desktop\sungyeh\python\2019-03-26\MyFigure.png',dpi=300)
```
