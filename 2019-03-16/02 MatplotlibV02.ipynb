# 繪製圖格

```python
x=np.linspace(-3,3,50)
y1=3*x+3
y2=x**2
y3=np.random.randn(50)

# 設置畫布
fig=plt.figure(num=2,figsize=(12,8))

# 設置圖框
ax1=fig.add_subplot(2,3,1)
ax2=fig.add_subplot(2,3,2)
ax4=fig.add_subplot(2,3,4)

# 畫布標題
fig.suptitle('My Figure') # Note: "p" in "suptitle"

# 圖框標題
ax1.set_title('ax1')
ax2.set_title('ax2')
ax4.set_title('ax4')

# 繪圖
ax1.plot(x,y1,label='ya')
ax2.plot(x,y2,'r--')
ax4.scatter(x,y3)

# 繪製圖例
ax1.legend()

# 調整圖框座標
ax1.set_xlim(-3,3)

# 設置圖框軸標題
ax2.set_xlabel('this is x')
ax2.set_ylabel('this is y')
```

# 直方圖 Histogram
### 驗證 "中央極限定理"

```python
plt.figure()
np.random.seed(10)
x=np.random.randint(105,886,1000)

# 3 random generator have been learned

# np.arange(start,end,step)
# np.linspace(start,end,intervals)
# np.random.randint(start,end,items)

plt.subplot(1,2,1)
plt.hist(x,edgecolor='white')
plt.title('raw data')

# 抽樣次數
t=1000
# 每次抽樣個數
n=25

xs=[]
for i in range(t):
    a=np.random.choice(x,n,replace=True)
    b=a.sum()/n
    xs.append(b)
    
plt.subplot(1,2,2)
plt.hist(xs,edgecolor='white')
plt.title('after sampling');
```
